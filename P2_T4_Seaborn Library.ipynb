```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
print(sns.relplot(x="passengers",y="month",data=a))
plt.show()
```

```python
a = sns.load_dataset("flights")
print(sns.relplot(x="passengers",y="month",hue="year",data=a))
plt.show()
```

```python
# Visualizing Statical Relationships
# Process of understanding relationships between
# variables of a dataset
```

```python
b = sns.load_dataset('tips')
print(b)
```

```python
print(sns.relplot(x='time',y='tip',data=b,kind='line'))
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=b)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',hue='xender',data=b)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='violin')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='violin',hue='sex')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='boxen')
plt.show()
```

# Displot

```python
# Distribution of a univariate set of observations
sns.displot(b['tip'],kde=1)
```

# Multi-Plot grid

```python
a = sns.load_dataset('iris')
b = sns.FacetGrid(a,col='species')
b.map(plt.hist,'sepal_length')
plt.show()
```

```python
sns.set(style='darkgrid')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
plt.show()
```

```python
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
plt.show()
```

```python
sns.set(style='white',color_codes=True)
b=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=b)
plt.show()
```

```python
sns.set(style='white',color_codes=True)
b=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=b)
sns.despine(offset=10,trim=True)
plt.show()
```

```python
c=sns.color_palette()
sns.palplot(c)
plt.show()
```
