```python
# Loding the library with the iris dataset
from sklearn.datasets import load_iris
# Loding scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
#setting random seed
np.random.seed(0)
```

```python
# Creating an object called iris with the iris data
iris=load_iris()
#create dataframe
df=pd.DataFrame(iris.data,columns=iris.feature_names)
#Viewing the top 5 rows
df.head()
```

```python
#Adding a new column for the species name
df["species"]=pd.Categorical.from_codes(iris.target,iris.target_names)
df
```

```python
#creating Test and train data
df["is_train"]=np.random.uniform(0,1,len(df))<=0.75
#View data
df
```

```python
#creating dataframe with test rows and training rows
train,test=df[df["is_train"]==True],df[df["is_train"]==False]
#show the number of observation for the test and training dataframe
print("No. of observation in training data:",len(train))
print("No. of observation in the test data:",len(test))
```

```python
#Create a list of the feature column's names
features=df.columns[:4]
#viewing features
features
```

```python
#converting each species nam into digits
y=pd.factorize(train["species"])[0]
#viewing target
y
```

```python
help(RandomForestClassifier)
```

```python
#creating a random forest Classifier
clf=RandomForestClassifier(n_jobs=2,random_state=0)
#training the classifier
clf.fit(train[features],y)
```

```python
#Applying the trained classifier to the test
clf.predict(test[features])
```

```python
features
```

```python
#viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
clf.predict_proba(test[features])[10:20]
```

```python
#mapping names for the plants for each predicted plant class
preds=iris.target_names[clf.predict(test[features])]
#view the predicted species for the first five observation
preds[0:5]
```

```python
#fusion matrix
pd.crosstab(test["species"],preds,rownames=["Actual Species"],colnames=["Predicted Species"])
```
