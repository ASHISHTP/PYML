### Legends,Title and Label

```python
x=[1,2,3,4]
y=[5,7,4,6]

x2=[1,2,3,4]
y2=[10,14,12,13]

plt.plot(x,y,label='Loss Per Month',color='R')
plt.plot(x2,y2,label='Profit Per Month',color='B')

plt.xlabel('Months',fontsize='xx-large')
plt.ylabel('Lakhs in Rupees',fontsize='xx-large')
plt.title('XYZ Company \n Loss-Profit',fontweight='bold')
plt.legend()
plt.show()
```

### Bar Charts

```python
x=[1,3,5,7]
y=[5,7,4,6]

x2=[2,4,6,8]
y2=[10,14,12,13]

plt.bar(x,y,label='Loss Per Month',color='R')
plt.bar(x2,y2,label='Profit Per Month',color='B')

plt.xlabel('Months',fontsize='xx-large')
plt.ylabel('Lakhs in Rupees',fontsize='xx-large')
plt.title('XYZ Company \n Loss-Profit',fontweight='bold')
plt.legend()
plt.show()
```

### StackPlot

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','k','b','r'],labels=['Sleeping','Eating','Working','Playing'])

plt.xlabel('Days')
plt.ylabel('Total hours spend')
plt.title('Stack of Day Time utilization')
plt.legend()
plt.show()
```

### Pie Chart

```python
slices=[7,2,8,7]
activities=['Sleeping','Eating','Working','Playing']
cols=['c','m','r','g']
plt.pie(slices,labels=activities,colors=cols,startangle=120,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%',radius=0.8)
plt.title('Monday Time Utilization')
plt.legend()
plt.show()
```
