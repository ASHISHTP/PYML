```python
from scipy import integrate
import numpy as np
```

# Integration

```python
x2 = lambda x: x**2
print(integrate.quad(x2,0,4))
#Here function returns two values, in which the first value is integration
#and second value is estimated error in integrtion.
print(4**3/3)
```

```python
#int(e^-x) = (-e^-x)
invexp=lambda x: np.exp(-x)
print(integrate.quad(invexp,0,np.inf))
a,b=(integrate.quad(invexp,0,np.inf))
print(a)
print(b)
```

```python
invsin=lambda x: np.sin(x)
print(integrate.quad(invsin,0,np.pi))
```

# Double Integration

```python
#import square root function from mat lib
from math import sqrt
#set funcytion f(x)
f = lambda x,y: 64*x*y
# lower limit of second integral 
p = lambda x: 0
# upper limit of first integral
q = lambda y: sqrt(1 - 2*y**2)
# perform double integration
integration = integrate.dblquad(f,0,2/4,p,q)
print(integration)
```

# Optimization

```python
from scipy import optimize
from matplotlib import pyplot as plt
```

```python
################## Finding the minimum of scalar function ####################
def f(x):
    return x**2 + 10*np.sin(x)
x = np.arange(-10,10,0.1)
plt.plot(x,f(x))
plt.show()
```

```python
#################### BFGS Algorithm #######################
print(optimize.fmin_bfgs(f,3))
```

```python
print(optimize.fmin_bfgs(f,0))
```

```python
################### basinhopping ######################
print(optimize.basinhopping(f,0))
```

# Scipy with Images

```python
from scipy import misc
from matplotlib import pyplot as plt
import numpy as np
# get face image from misc package
panda=misc.face()
# plot or show image face
plt.imshow(panda)
plt.show()
```

```python
panda.shape
```

```python
plt.imshow(panda.reshape(1024,768,3))
plt.show()
```

```python
panda.ndim
```

```python
from scipy import ndimage, misc
# rotation function of scipy for image rotated 135 degree
panda_rotate=ndimage.rotate(panda,135)
plt.imshow(panda_rotate)
plt.show()
```

```python
from scipy import ndimage, misc
# rotation function of scipy for image rotated 90 degree
panda_rotate=ndimage.rotate(panda,90)
plt.imshow(panda_rotate)
plt.show()
```
