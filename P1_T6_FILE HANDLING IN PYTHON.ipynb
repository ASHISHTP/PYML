# Read any file

```python
fobj = open("Example.txt","r")
for line in fobj:
    print(line.rstrip())
fobj.close()
```

```python
fobj = open("Example.txt","r")
for line in fobj:
    print((line.upper()).rstrip())
fobj.close()
```

# With clause

```python
with open('Example_Write_With.txt','w') as fh:
    fh.write("There are 12 months\nIn one calander year")
```

```python
with open("Example.txt") as fobj:
    for line in fobj:
        print(line.rstrip())
```

```python
Sum=0
with open('number.txt') as num:
    for i in num:
        Sum=Sum+int(i)
    print(Sum)
```

```python
with open("prime.txt","a") as pr:
    for i in range(2,201,1):
        for j in range(2,i+1,1):
            if i%j==0:
                break
        if j==i:
            print(i)
            pr.write("{}\n".format(i))

```

# Pickle and Unpickle

```python
import pickle
```

```python
tour={"City":["Paris","New delhi","Mumbai","Nagpur"],"Hotels":["The Taj","Radison blu","Novalet"]}
fh=open("data.pkl","bw")
pickle.dump(tour,fh)
fh.close()
```

```python
f=open("data.pkl","rb")
new_tour=pickle.load(f)
print(new_tour)
f.close()
```

```python
f=open("data.pkl","rb")
new_tour=pickle.load(f)
for pointer in new_tour:
    print(new_tour[pointer])
f.close()
```
